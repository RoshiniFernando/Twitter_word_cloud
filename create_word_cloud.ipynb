```python
# Important libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import imageio.v2 as im
```

```python
# Creating DataFrame by reading .csv file 
python_tweets = pd.read_csv('/Users/roshinifer/Documents/Development/Twitter_word_cloud/my_python_tweets.csv')
```

```python
# Display Head
python_tweets.head()
```

```python
# Join my python tweets to a single string
words = ' '.join(python_tweets['tweet'])

# Remove URLs, RTs, and twitter handles
no_urls_no_tags = " ".join([word for word in words.split() 
                            if 'http' not in word 
                                and not word.startswith('@')
                                and word != 'RT'
                            ])

# Creating the wordcloud
wordcloud = WordCloud(                      
                      stopwords=STOPWORDS,
                      background_color='black',
                      width=1800,
                      height=1400
                     ).generate(no_urls_no_tags)

# Plot the wordcloud
plt.imshow(wordcloud)
plt.axis('off')
#plt.savefig('./my_twitter_wordcloud_1.png', dpi=300)
plt.show()
```

```python
# Twitter logo mask
twitter_mask = im.imread('./twitter_mask.png')

# Creating the wordcloud with the mask 
wordcloud = WordCloud( 
                      stopwords=STOPWORDS,
                      background_color='black',
                      width=1800,
                      height=1400,
                      mask=twitter_mask
            ).generate(no_urls_no_tags)

# Plot the wordcloud
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig('./my_twitter_wordcloud.png', dpi=300)
plt.show()
```
